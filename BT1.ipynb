a) Vẽ biểu đồ quạt (pie chart) cho nhóm A.

b) Vẽ biểu đồ thanh (bar chart) cho nhóm B.

c) Vẽ biểu đồ so sánh 2 nhóm.

```python
x = ['White bread', 'Sugar','Potatoes', 'Fruit', 'Vegetables', 'Brown bread'];
#a
y_A = [26,11.5,48.5,13,21.5,5]
plt.pie(y_A, labels = x, autopct="%.0f%%");
plt.title('Biểu đồ chế độ dinh dưỡng nhóm A');
plt.show();
```

```python
#b
y_B = [12.5,8,33.5,25.5,30.5,8]
plt.bar(x, y_B);
plt.title('Biểu đồ chế độ dinh dưỡng nhóm B');
plt.show();
```

```python
#c
height = 0.4;
y_pos = range(len(x));
y_pos1 = [y - height/2 for y in y_pos];
y_pos2 = [y + height/2 for y in y_pos];
plt.barh(y_pos1, y_A, height=height, label="Nhóm A");
plt.barh(y_pos2, y_B, height=height, label="Nhóm B");

plt.title("Chế độ dinh dưỡng");
plt.xlabel("Nhu cầu");
plt.xticks([10*i for i in range(7)]);
plt.yticks(range(len(x)), x);
plt.gca().invert_yaxis();
plt.legend();
plt.show();
```

Câu 3. Sau đây là số liệu điểm thi của một lớp (tính theo thang 20)

2, 3, 7, 7, 8, 8, 9, 10, 11, 11, 12, 12, 14, 14, 15, 17, 17, 18, 19, 20

a) Tính trung bình (mean), trung vị (median), yếu vị (mode), phân vị (percentile) 10%, 90%.

b) Tính phạm vi (range), khoảng tứ phân vị (IQR - interquartile range), phương sai (variance), độ lệch 
chuẩn (standard deviation).

c) Nhận xét.

```python
diemThi = [2, 3, 7, 7, 8, 8, 9, 10, 11, 11, 12, 12, 14, 14, 15, 17, 17, 18, 19, 20];
#a
trungBinh = np.mean(diemThi);
trungVi = np.median(diemThi);
yeuVi = stats.mode(diemThi)[0][0];
phanVi = np.percentile(diemThi, [10, 90]);
print('Trung Bình', trungBinh);
print('Trung Vị', trungVi);
print('Yếu Vị', yeuVi);
print('Phân Vị', phanVi);
```

```python
#b
phamVi = max(diemThi) - min(diemThi);
lower_quartile, upper_quartile = np.percentile(diemThi, [25, 75])
IQR = upper_quartile - lower_quartile
mean = sum(diemThi) / len(diemThi)
sum_square_deviation = sum((x - mean)**2 for x in diemThi)
var = sum_square_deviation / len(diemThi)
std = var ** 0.5
print('Phạm Vi:', phamVi);
print('Khoảng Tứ Phân Vị:', IQR);
print('Phương Sai:', var);
print('Độ Lệch Chuẩn:', std);
```

c/ Nhận xét

Điểm thi trung bình là 11.5

Điểm thi có phạm vi lớn 18.
